```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("diet_dataset.csv")
```

```python
df.isnull().sum()

plt.figure(figsize=(6,4))
plt.bar(df.columns, df.isnull().sum())
plt.xticks(rotation=90)
plt.title("Missing Values per Feature")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df["bmi"], bins=30)
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df["daily_calories"], bins=30)
plt.title("Daily Calories Distribution")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df["protein_intake"], df["carb_intake"])
plt.title("Protein vs Carbohydrates Intake")
plt.xlabel("Protein (g)")
plt.ylabel("Carbs (g)")
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(df["fat_intake"], df["protein_intake"])
plt.title("Fat vs Protein Intake")
plt.xlabel("Fat (g)")
plt.ylabel("Protein (g)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df["sleep_hours"], df["daily_steps"])
plt.title("Sleep Hours vs Daily Steps")
plt.xlabel("Sleep Hours")
plt.ylabel("Daily Steps")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df["daily_calories"], df["daily_steps"], s=10)
plt.title("Daily Calories vs Daily Steps")
plt.xlabel("Calories")
plt.ylabel("Steps")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure(figsize=(7,5))
plt.imshow(corr, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["diet_plan"].value_counts().plot(kind="bar")
plt.title("Diet Plan Class Distribution")
plt.xlabel("Diet Plan Type")
plt.ylabel("Count")
plt.grid(True, alpha=0.3)
plt.show()
```
